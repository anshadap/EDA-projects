```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df
```

```python
df.info()
```

```python
df=df.drop(columns='Status')
```

```python
df=df.drop(columns='unnamed1')
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
df.isna().sum()
```

```python
df.Amount.fillna(df.Amount.mean(),inplace=True)
```

```python
ax=sns.countplot(x=df.Gender)
ax.bar_label(ax.containers[0]);
```

```python
a=df.groupby('Gender').agg({'Amount':'sum'}).reset_index()
```

```python
sns.barplot(x=a.Gender,y=a.Amount)
```

from this graoh we can understand that female spend more amount than male for shoping  

```python
ax=sns.countplot(x=df['Age Group'],hue=df.Gender)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1]);
```

```python
b=df.groupby('Age Group').agg({'Amount':'sum'}).reset_index()
```

```python
sns.barplot(x=b['Age Group'],y=b.Amount)
```

most of the buyers are between 26-35

```python
c=df.groupby('State').agg({'Orders':'sum'}).sort_values(by='Orders',ascending=False)
```

```python
c=c.head(10).reset_index()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=c['State'],y=c['Orders'])
plt.xticks(rotation=45);
```

UP have good number of orders 

```python
d=df.groupby('State').agg({'Amount':'sum'}).sort_values(by='Amount',ascending=False)
d=d.head(10).reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x=d['State'],y=d['Amount'])
plt.xticks(rotation=45);
```

here you can see keala is not there because kerela have low purchasing power

```python
sns.countplot(x=df['Marital_Status'],hue=df.Gender)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sns.countplot(x=df.Occupation)
```

```python
e=df.groupby('Occupation').agg({"Amount":'sum'}).sort_values(by='Amount',ascending=False).reset_index()
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x=e.Occupation,y=e.Amount)
```

```python
plt.figure(figsize=(20,8))
ax=sns.countplot(x=df['Product_Category'])
ax.bar_label(ax.containers[0])
plt.xticks(rotation=45);
```

```python
z=df.pivot_table(index='Age Group',columns='Product_Category',values='Amount',aggfunc='mean')
```
